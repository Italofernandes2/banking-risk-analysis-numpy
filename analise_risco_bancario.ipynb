```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Garantindo reprodutibilidade
np.random.seed(42)

# Gerando dados fictícios
id_clientes = np.arange(1, 1001)
renda_mensal = np.random.uniform(1500, 20000, 1000)
saldo_conta = np.random.uniform(0, 1_000_000, 1000)
cartao_credito = np.random.uniform(0, 15000, 1000)
investimentos = np.random.uniform(0, 10_000_000, 1000)
atrasos = np.random.randint(0, 90, 1000)

# Criando o DataFrame
df = pd.DataFrame({
    'id_cliente':   id_clientes,
    'renda_mensal': renda_mensal,
    'saldo_conta':  saldo_conta,
    'cartao_credito': cartao_credito,
    'investimentos':  investimentos,
    'atrasos_dias':   atrasos
})

# Criando os segmentos de clientes
df['segmento'] = np.where(df['atrasos_dias'] > 30, 'Alto Risco', 'Baixo Risco')

#Analise estatistica
print("=== PERFIL MÉDIO DOS CLIENTES ===")
print(f"Renda Mensal:   Média R$ {np.mean(renda_mensal):.2f} | Mediana R$ {np.median(renda_mensal):.2f} | Desvio Padrão R$ {np.std(renda_mensal):.2f}")
print(f"Saldo Conta:    Média R$ {np.mean(saldo_conta):.2f} | Mediana R$ {np.median(saldo_conta):.2f} | Desvio Padrão R$ {np.std(saldo_conta):.2f}")
print(f"Investimentos:  Média R$ {np.mean(investimentos):,.2f} | Mediana R$ {np.median(investimentos):.2f} | Desvio Padrão R$ {np.std(investimentos):.2f}")
print(f"Atrasos (dias): Média {np.mean(atrasos):.1f} | Mediana {np.median(atrasos):.1f} | Desvio Padrão {np.std(atrasos):.1f}")

# Separando os segmentos
alto_risco = df[df['segmento'] == 'Alto Risco']
baixo_risco = df[df['segmento'] == 'Baixo Risco']

# Comparando os segmentos
print("=== ALTO RISCO ===")
print(f"Quantidade de clientes: {len(alto_risco)}")
print(f"Renda Média:    R$ {np.mean(alto_risco['renda_mensal']):,.2f}")
print(f"Saldo Médio:    R$ {np.mean(alto_risco['saldo_conta']):,.2f}")
print(f"Investimentos:  R$ {np.mean(alto_risco['investimentos']):,.2f}")
print(f"Atraso Médio:   {np.mean(alto_risco['atrasos_dias']):.1f} dias")

print("\n=== BAIXO RISCO ===")
print(f"Quantidade de clientes: {len(baixo_risco)}")
print(f"Renda Média:    R$ {np.mean(baixo_risco['renda_mensal']):,.2f}")
print(f"Saldo Médio:    R$ {np.mean(baixo_risco['saldo_conta']):,.2f}")
print(f"Investimentos:  R$ {np.mean(baixo_risco['investimentos']):,.2f}")
print(f"Atraso Médio:   {np.mean(baixo_risco['atrasos_dias']):.1f} dias")

# Calculando correlações
print("=== CORRELAÇÕES ===")

corr_renda_invest = np.corrcoef(df['renda_mensal'], df['investimentos'])[0,1]
corr_renda_saldo = np.corrcoef(df['renda_mensal'], df['saldo_conta'])[0,1]
corr_atraso_saldo = np.corrcoef(df['atrasos_dias'], df['saldo_conta'])[0,1]

print(f"Renda x Investimentos: {corr_renda_invest:.2f}")
print(f"Renda x Saldo Conta:   {corr_renda_saldo:.2f}")
print(f"Atraso x Saldo Conta:  {corr_atraso_saldo:.2f}")

# Gráfico 1 - Distribuição da Renda por Segmento
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='renda_mensal', hue='segmento', bins=30)
plt.title('Distribuição da Renda Mensal por Segmento')
plt.xlabel('Renda Mensal (R$)')
plt.ylabel('Quantidade de Clientes')
plt.show()
plt.savefig('distribuicao_renda.png')

# Gráfico 2 - Matriz de Correlação
plt.figure(figsize=(8, 6))
correlacoes = np.corrcoef([df['renda_mensal'], df['saldo_conta'], 
                           df['investimentos'], df['atrasos_dias']])
sns.heatmap(correlacoes, annot=True, fmt='.2f',
            xticklabels=['Renda', 'Saldo', 'Investimentos', 'Atrasos'],
            yticklabels=['Renda', 'Saldo', 'Investimentos', 'Atrasos'])
plt.title('Matriz de Correlação')
plt.show()
plt.savefig('matriz_correlacao.png')
```
